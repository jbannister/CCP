# What is coding?
To tell a computer what to do, you generally need to write a computer program. 

The vast majority of programming languages are designed to be used by humans to write instructions that computers can execute. 

The first programming languages predate the modern computer, and were used to direct the behavior of machines such as Jacquard looms and player pianos. Thousands of different programming languages have been created.

### Scratch
Scratch is a visual programming language, meaning that its code is made up of colorful blocks that snap together to make scripts. It is easy to use, free, and runs in a web browser.

Scratch is designed to be fun. It is easy to get started, but also powerful enough to create complex animations and games.

Scratch was originally developed by the Lifelong Kindergarten Group at the MIT Media Lab, and has been used to teach millions of kids around the world how to code.

The Scratch Team constantly improves Scratch, and releases new versions every few months. 

### How Scratch works
Programs are made by placing a sequence of blocks on a grid. The programmer can observe the results of their code as it executes, as it is being written.

### A typical Scratch project 
Scratch can be used to create games, animations, stories, art, and more.

In Scratch, users can program their own interactive stories, games, and animations, by snapping together graphical programming blocks. The Scratch programming language is based on the concept of "blocks" that can be dragged and dropped together to create scripts.
